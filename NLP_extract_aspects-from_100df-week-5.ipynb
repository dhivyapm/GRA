```python
def aspect_sentiment_analysis(txt,nlp):
    
    txt = txt.lower() # LowerCasing the given Text
    sentList = nltk.sent_tokenize(txt) # Splitting the text into sentences

    fcluster = []
    feature_list = []
    dic = {}

    for line in sentList:
        new_txt_list = nltk.word_tokenize(line)
        taggedList = nltk.pos_tag(new_txt_list)

        doc = nlp(line) # Object of Stanford NLP Pipeleine
        
        # Getting the dependency relations betwwen the words
    #print(taggedList)
    dep_node = []
    for dep_edge in doc.sentences[0].dependencies:
        dep_node.append([dep_edge[2].text, dep_edge[0].text, dep_edge[1]])
    #print(dep_node)
    categoryList = []
    featureList = []
    for i in range(0,len(taggedList)-1):
        if(taggedList[i][1]=="JJ" and taggedList[i+1][1]=="NN"):
            featureList.append(taggedList[i])
            categoryList.append(taggedList[i][0])
            featureList.append(taggedList[i+1])
            categoryList.append(taggedList[i+1][0])
    #print('feature list:',featureList,'\n')
    #print('category list:',categoryList)

    for i in featureList:
        filist = []
        for j in dep_node:
            if((j[0]==i[0] or j[1]==i[0]) and (j[2] in ["nsubj", "obj","advmod", "amod","compound"])):
                if(j[0]==i[0]):
                    filist.append(j[1])
                else:
                    filist.append(j[0])
        fcluster.append(str(i[0])+str(filist))
        feature_list.append(i[0])
        
    return(feature_list) 
```

```python
nlp = stanza.Pipeline()
```

```python
cluster = []
for i in range(0,100):
    cluster.append(aspect_sentiment_analysis(df['clean'][i],nlp))
```

```python
def flatten(seq):
    for item in seq:
        if isinstance(item, list):
            yield from flatten(item)
        else:
            yield item
```

```python
result = []
for i in range(0, 100):
    result.extend(flatten(cluster[i]))
```

```python
from collections import Counter
def remove_words(lst, k): 
    counted = Counter(lst) 
    #print(counted)  
    temp_lst = [] 
    for el in counted: 
        if counted[el] < k: 
            temp_lst.append(el) 
    #print(temp_lst)          
    res_lst = [] 
    for el in lst: 
        if el not in temp_lst: 
            res_lst.append(el) 
              
    return(res_lst)
```

```python
freq_feature = remove_words(result,15)
```
