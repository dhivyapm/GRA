```python
import pandas as pd
import nltk
import sys
```

```python
df = pd.read_csv('movie_emotions.csv',sep=',')
data_new = df.copy()
```

```python
def aspect_VN(txt):
    
    sentList = nltk.sent_tokenize(txt) # Splitting the text into sentences

    for line in sentList:
        new_txt_list = nltk.word_tokenize(line)
        taggedList = nltk.pos_tag(new_txt_list)

  
    categoryList = []

    for i in range(0,len(taggedList)-1):
        if((taggedList[i][1]=="VB" or taggedList[i][1]=="VBP") and (taggedList[i+1][1]=="NN" or taggedList[i+1][1]=="NNS" or taggedList[i+1][1]=="NNP")):

            categoryList.append(taggedList[i][0])

            categoryList.append(taggedList[i+1][0])


    return categoryList
```

```python
import stanza
stanza.download('en')
nlp = stanza.Pipeline()
```

```python
aspect_cluster = [aspect_VN(data_new['unique_sent'][unt]) for unt in range(0,10)]
```

```python
aspect = []
for term in aspect_cluster:
    _term = ''
    for word in term:
        _term += ' ' + word
    aspect.append(_term.strip())
aspect
```

```python
import text2emotion as te
import numpy as np
def text_2_emotions(text):
    r = []
    result = te.get_emotion(text)
    for k,v in result.items(): 
        if result[k] != 0.0:
            r.append(max(result, key=result.get))
    return np.unique(r)
res = []
for asp in aspect:
    res.append(text_2_emotions(str(asp)))  
```

```python
aspect_data = pd.DataFrame()
```

```python
aspect_data['aspects'] = aspect
```

```python
aspect_data['emotions'] = res
```

```python
aspect_data.to_csv('aspect_analysis_1.csv')
```
